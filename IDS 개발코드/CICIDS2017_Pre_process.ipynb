## IDS Data-set Pre-Processing

CICIDS2017 데이터셋 전처리

### 데이터 전처리

```python
import pandas as pd # Use pandas lib
```

```python
# dataset 불러오기
dataset =pd.read_csv('./TrafficLabelling/Wednesday-workingHours.pcap_ISCX.csv')

# 데이터 is NaN 제거
dataset = dataset.dropna(axis=0) # 데이터 is NaN 제거

dataset
```

```python
dataset.columns
```

```python
# 추출할 컬럼 리스트
cols_to_keep = [' Source IP', ' Destination IP', ' Protocol', ' Source Port', ' Destination Port', ' Flow Duration',
                'FIN Flag Count', ' SYN Flag Count', ' RST Flag Count', ' PSH Flag Count', ' ACK Flag Count',
                ' URG Flag Count', ' CWE Flag Count', ' ECE Flag Count', 'Flow Bytes/s', ' Flow Packets/s',
                'Fwd Packets/s', ' Min Packet Length', ' Max Packet Length', ' Packet Length Mean',
                ' Packet Length Std', ' Packet Length Variance',' Label']

# 요구하는 컬럼만 추출하여 새로운 DataFrame 생성
new_dataset = dataset[cols_to_keep]
```

```python
new_dataset
```

```python
new_dataset = new_dataset.rename(columns={' Source IP':'Source IP', ' Destination IP':'Destination IP', ' Protocol':'Protocol', ' Source Port':'Source Port',
                                  ' Destination Port':'Destination Port', ' Flow Duration':'Flow Duration',
                                  ' SYN Flag Count':'SYN Flag Count', ' RST Flag Count':'RST Flag Count', ' PSH Flag Count':'PSH Flag Count',
                                  ' ACK Flag Count':'ACK Flag Count',' URG Flag Count':'URG Flag Count', ' CWE Flag Count': 'CWE Flag Count',
                                  ' ECE Flag Count':'ECE Flag Count',' Flow Packets/s':'Flow Packets/s',' Min Packet Length':'Min Packet Length',
                                  ' Max Packet Length':'Max Packet Length', ' Packet Length Mean':'Packet Length Mean',' Packet Length Std':'Packet Length Std',
                                  ' Packet Length Variance':'Packet Length Variance',' Label':'Label'})

new_dataset.columns
```

```python
new_dataset
```

```python
protocol_counts = new_dataset['Label'].value_counts()
print(protocol_counts)
```

```python
#트래픽 분류 Normal : 0 , DDos : 1 , PortScan : 2 , Botnet : 3 , Infiltration : 4 , FTP-Patator : 5
# SSH-Patator : 6, DoS Hulk : 7 , DoS GoldenEye : 8 , DoS slowloris : 9 , DoS Slowhttptest : 10
# Heartbleed : 11
new_dataset['Label'] = new_dataset['Label'].apply(lambda x: 0 if x == 'BENIGN' else 1 if x == 'DDoS' else x)
new_dataset['Label'] = new_dataset['Label'].apply(lambda x: 2 if x == 'PortScan' else 3 if x == 'Bot' else x)
new_dataset['Label'] = new_dataset['Label'].apply(lambda x: 4 if x == 'Infiltration' else 5 if x == 'FTP-Patator' else x)
new_dataset['Label'] = new_dataset['Label'].apply(lambda x: 6 if x == 'SSH-Patator' else 7 if x == 'DoS Hulk' else x)
new_dataset['Label'] = new_dataset['Label'].apply(lambda x: 8 if x == 'DoS GoldenEye' else 9 if x == 'DoS slowloris' else x)
new_dataset['Label'] = new_dataset['Label'].apply(lambda x: 10 if x == 'DoS Slowhttptest' else 11 if x == 'Heartbleed' else x)

new_dataset
```

```python
new_dataset.to_csv('DoS_Hulk_GoldenEye_slowloris_Slowhttptest_Heartbleed_Normal_data.csv', index=False)
```
